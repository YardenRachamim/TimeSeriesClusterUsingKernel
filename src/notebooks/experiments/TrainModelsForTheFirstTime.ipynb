```python
from utils import DataUtils
from TCK import TCK
from GMM_MAP_EM import GMM_MAP_EM

from pathlib import Path
import numpy as np
import pickle

from sklearn.metrics import accuracy_score
from sklearn.decomposition import KernelPCA
import matplotlib.pyplot as plt
```

```python
root_path = Path(r"..\..\..\.")
data_path = root_path / "data"
models_path = root_path / "models"
```

```python
similarity_function = "jensenshannon"
C = 40
Q = 30 
single_gmm_num_iter = 20
max_features = 'all'
```

```python
dataset_name = "PenDigits"
model_name = f"similarity_function-{similarity_function}_C-{C}_Q-{Q}_single_gmm_num_iter-{single_gmm_num_iter}_max_features-{max_features}"
```

```python
path = models_path / dataset_name / model_name
path.mkdir(parents=True, exist_ok=True)
```

# Train for the first time

### Load the numpy arrays

note that they are preprocessed before

```python
X_train, X_test, y_train, y_test = DataUtils.load_numpy_array(data_path / dataset_name)
```

```python
R_train = (~(np.isnan(X_train))).astype(int)
R_test = (~(np.isnan(X_test))).astype(int)
```

### Set the model

```python
tck_model = TCK(
    Q=Q,
    C=C,
    n_jobs=6,
    max_features=max_features,
    similarity_function=similarity_function,
    single_gmm_num_iter=single_gmm_num_iter
)
```

### Train the model

```python
tck_model.fit(X_train, R_train)
```

### Get K star

```python
K_star = tck_model.transform(X_test, R_test)
```

### Save the model

```python
with open(path / "model", "wb") as fos:
    pickle.dump(tck_model, fos)
```

```python
np.save(path / "K_Star", K_star)
```
